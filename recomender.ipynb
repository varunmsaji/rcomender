```python
 #popularity based recomender 
#min 250 rating and top avg rating 
#we need to merge rating and user using isbin col
dfp = rating.merge(books,on='ISBN')
```

```python
dfp.shape
```

```python
           
num_rating_df=dfp.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
def pre(txt):
    txt = re.sub('[^a-zA-Z]',' ',txt)
    return txt
```

```python
num_rating_df['Book-Title'] = num_rating_df['Book-Title'].apply(pre)
```

```python
num_rating_df
```

```python
           
avg_df=dfp.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_df
```

```python
popular_df = num_rating_df.merge(avg_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=True)
```

```python
popular_df
```

```python
popular_df.shape
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
#colaberative filtering we use dfp
dfp.head()
```

```python
x = dfp.groupby('User-ID').count()['Book-Rating']>200
rateduser = x[x].index
```

```python
filtered_user = dfp[dfp['User-ID'].isin(rateduser)]
```

```python
filtered_user
#we have all the users who have rated more than 200 books
```

```python
y = filtered_user.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
finale_ratings = filtered_user[filtered_user['Book-Title'].isin(famous_books)]
```

```python
finale_ratings
```

```python
finale_ratings.drop_duplicates()
```

```python
pt = finale_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#now each books are like a vector andy now we need to calculate distance from each
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recoment(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
        
```

```python
def recoment(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    data =[]
    for i in similar_items:
        item =[]
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```
